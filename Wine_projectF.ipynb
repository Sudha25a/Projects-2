```python
# Scatter Plot
# Create a scatter plot of alcohol content vs. quality
plt.scatter(Redwine['alcohol'], Redwine['quality'], label='Red Wine')
plt.scatter(Whitewine['alcohol'], Whitewine['quality'], label='White Wine')
plt.xlabel('Quality')
plt.ylabel('Alcohol Content')
# plt.legend()
plt.show()
```

```python
# Bar Chart
# Create a bar chart of average quality by wine type
avg_quality_red =Redwine['quality'].mean()
avg_quality_white = Whitewine['quality'].mean()

plt.bar(['Red Wine', 'White Wine'], [avg_quality_red, avg_quality_white])
plt.xlabel('Wine Type')
plt.ylabel('Average Quality')
plt.show()
```

```python
# Histogram
# Create a histogram of acidity levels
plt.hist(Redwine['fixed acidity'], label='Red Wine', alpha=0.5)
plt.hist(Whitewine['fixed acidity'], label='White Wine', alpha=0.5)
plt.xlabel('Acidity Level')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Correlation Matrix
# Create a correlation matrix for red wine features
corr_matrix_red = Redwine.corr()
sns.heatmap(corr_matrix_red, annot=True, cmap='coolwarm', square=True)
plt.title('Red Wine Correlation Matrix')
plt.show()

# Create a correlation matrix for white wine features
corr_matrix_white = Whitewine.corr()
sns.heatmap(corr_matrix_white, annot=True, cmap='coolwarm', square=True)
plt.title('White Wine Correlation Matrix')
plt.show()
```
