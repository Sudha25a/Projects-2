```python
# Scatter Plot
# Create a scatter plot of alcohol content vs. quality
plt.scatter(Redwine['alcohol'], Redwine['quality'], label='Red Wine')
plt.scatter(Whitewine['alcohol'], Whitewine['quality'], label='White Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.legend()
plt.show()
```

```python
plt.scatter(Redwine['alcohol'], Redwine['quality'], label='Red Wine')
plt.scatter(Whitewine['alcohol'], Whitewine['quality'], label='White Wine')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.legend()
plt.show()
```

```python
# Bar Chart
# Create a bar chart of average quality by wine type
avg_quality_red =Redwine['quality'].mean()
avg_quality_white = Whitewine['quality'].mean()

plt.bar(['Red Wine', 'White Wine'], [avg_quality_red, avg_quality_white])
plt.xlabel('Wine Type')
plt.ylabel('Average Quality')
plt.show()
```

```python
# Histogram
# Create a histogram of acidity levels
plt.hist(Redwine['fixed acidity'], label='Red Wine', color= 'r', alpha=0.8)
plt.hist(Whitewine['fixed acidity'], label='White Wine',color='g', alpha=0.8)
plt.xlabel('Acidity Level')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Create a correlation matrix for red wine features
# corr_matrix_red = Redwine.corr()
# sns.heatmap(corr_matrix_red, cmap='coolwarm')
# plt.title('Red Wine Correlation Matrix')
# plt.show()

corr_matrix = Whitewine.corr()
sns.heatmap(corr_matrix, cmap ='coolwarm')
plt.title('White Wine Correlation Matrix')
plt.show()
```

```python
corr_matrix = Redwine.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()


corr_matrix = Whitewine.corr()
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.show()
```

```python
# Correlation Matrix
#Create a correlation matrix for red wine features
corr_matrix_red = Redwine.corr()
sns.heatmap(corr_matrix_red, cmap='coolwarm')
plt.title('Red Wine Correlation Matrix')
plt.show()

# Create a correlation matrix for white wine features
corr_matrix_white = Whitewine.corr()
sns.heatmap(corr_matrix_white, cmap='coolwarm')
plt.title('White Wine Correlation Matrix')
plt.show()
```

```python
# Concatenate the two dataframes
Redwine = pd.concat([Redwine.assign(type='Red')])
Whitewine =pd.concat([Whitewine.assign(type='White')])

sns.set_theme(style="darkgrid")

# Create a scatter plot
sns.lineplot(x='alcohol', y='quality', hue='type',palette='set1', data=Redwine)
sns.lineplot(x='alcohol', y='quality', hue='type',palette='set1', data=Whitewine)
# Set labels and title
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')

# Show the plot
plt.show()
```
