```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
Redwine = pd.read_csv("winequality-red.csv", delimiter=';')
Redwine
```

```python
Redwine.groupby('quality').mean()
```

```python
# Redwine['fixed acidity'].plot(kind = 'box')
sns.boxplot(data= Redwine, x='pH')
```

```python
Redwine.skew()
```

```python
# Create a correlation matrix for red wine features
plt.figure(figsize=(12,8))
corr_matrix_red = Redwine.corr()

# Create a mask for the upper triangle
# mask = np.triu(np.ones_like(corr_matrix_red, dtype=bool))
sns.heatmap(corr_matrix_red, annot=True, cmap='coolwarm', square=True)
# mask=mask
plt.title('Red Wine Correlation Matrix')
plt.show()
```

```python
plt.scatter(data=Redwine, x= 'alcohol', y = 'quality', label='Red Wine', color= 'r')
```

```python
#  Plot 1: Red wine pH vs. acidity
sns.scatterplot(x='pH', y='fixed acidity', data=Redwine)
```
