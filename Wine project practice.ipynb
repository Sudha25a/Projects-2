```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Redwine = pd.read_csv("winequality-red.csv", delimiter=';')
Redwine
```

```python
Whitewine = pd.read_csv("winequality-white.csv", delimiter=';')
Whitewine
```

```python
# Assuming Redwine and Whitewine are DataFrames with a 'quality' column

sns.set_style('whitegrid')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 9), sharey=True)

# Red wine quality count plot
sns.countplot(x='quality', data=Redwine, label='Red Wine', ax=axes[1], color='r')

# White wine quality count plot
sns.countplot(x='quality', data=Whitewine, label='White Wine', ax=axes[0], color='g')

# Titles for each plot
axes[1].set_title('Red Wine Quality', alpha=0.8)
axes[0].set_title('White Wine Quality', alpha=0.8)

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr_matrix = Whitewine.corr()
sns.heatmap(corr_matrix, cmap='coolwarm',annot=True, square=True)
```

```python
# Create a correlation matrix for red wine features
plt.figure(figsize=(12,8))
corr_matrix= Whitewine.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Plot the heatmap with the mask
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True, mask=mask)

# Set title
plt.title('White Wine Correlation Matrix')

# Display the plot
plt.show()
```

```python
# Create a scatter plot to visualize the relationship between pH and fixed acidity
plt.figure(figsize=(10,6))
sns.scatterplot(x='pH', y='fixed acidity', data=Redwine, label='Red Wine', color ='r')
sns.scatterplot(x='pH', y='fixed acidity', data=Whitewine, label='White Wine', color = 'g')
plt.title('Relationship between pH and Acidity')
plt.xlabel('pH')
plt.ylabel('fixed acidity')
plt.legend()
plt.show()
```

```python
# Create a histogram of acidity levels
plt.hist(Redwine['fixed acidity'], label='Red Wine', color= 'r', alpha=0.8)
plt.hist(Whitewine['fixed acidity'], label='White Wine',color='g', alpha=0.8)
plt.title('Comparative Acidity Analysis')
plt.xlabel('Acidity Level')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Create a histogram of acidity levels
sns.set_style('whitegrid')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 9), sharey=True)
# plt.hist(Redwine['fixed acidity'], label='Red Wine', color= 'r', alpha=0.8)
# plt.hist(Whitewine['fixed acidity'], label='White Wine',color='g', alpha=0.8)
# plt.title('Comparative Acidity Analysis')
# plt.xlabel('Acidity Level')
# plt.ylabel('Frequency')
# plt.legend()
# plt.show()

# Red wine fixed acidity plot
sns.countplot(x= 'fixed acidity', data=Redwine, label='Red Wine', ax=axes[1], color='r')

# White wine fixed acidity plot
sns.countplot(x='fixed acidity', data=Whitewine, label='White Wine', ax=axes[0], color='g')

# Titles for each plot
axes[1].set_title('Red Wine Quality', alpha=0.8)
axes[0].set_title('White Wine Quality', alpha=0.8)

# Display the plot
plt.show()
```

```python
# Create a box plot to compare the distribution of fixed acidity in red and white wines
plt.figure(figsize=(10,6))
sns.boxplot(x='wine_type', y='fixed acidity', data=pd.concat([Redwine.assign(wine_type='Red'), Whitewine.assign(wine_type='White')]))
plt.title('Comparison of Acidity Distribution')
plt.xlabel('Wine Type')
plt.ylabel('Acidity')
plt.show()
```

```python
# Create a violin plot to compare the distribution of pH in red and white wines
plt.figure(figsize=(12,8))
sns.catplot(x='wine_type', y='pH', data=pd.concat([Redwine.assign(wine_type='Red'), Whitewine.assign(wine_type='White')]))
plt.title('Comparison of pH Distribution')
plt.xlabel('Wine Type')
plt.ylabel('pH')
plt.show()
```

```python
variable = 'residual sugar'
sns.kdeplot(Redwine[variable])
# sns.kdeplot(Redwine[variable], shade=True)
sns.kdeplot(Whitewine[variable])
# sns.kdeplot(Whitewine[variable], shade=True)

plt.title(f'Density Plot of {variable}')
plt.xlabel(variable)
plt.ylabel('Density')

# Show the plot
plt.show()
```
