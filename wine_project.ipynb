```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
Redwine = pd.read_csv("winequality-red.csv", delimiter=';')
Redwine
```

```python
Whitewine = pd.read_csv("winequality-white.csv", delimiter=';')
Whitewine
```

```python
# Create a scatter plot to visualize the relationship between pH and fixed acidity
plt.figure(figsize=(10,6))
sns.scatterplot(x='pH', y='fixed acidity', data=Redwine, label='Red Wine')
sns.scatterplot(x='pH', y='fixed acidity', data=Whitewine, label='White Wine')
plt.title('Relationship between pH and Acidity')
plt.xlabel('pH')
plt.ylabel('fixed acidity')
plt.legend()
plt.show()
```

```python
# Histogram
# Create a histogram of acidity levels
plt.hist(Redwine['fixed acidity'], label='Red Wine', color= 'r', alpha=0.8)
plt.hist(Whitewine['fixed acidity'], label='White Wine',color='g', alpha=0.8)
plt.title('Comparative Acidity Analysis')
plt.xlabel('Acidity Level')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Create a bar plot to compare the quality of red and white wines
plt.figure(figsize=(10,6))
sns.countplot(x='quality', data=Redwine, label='Red Wine')
sns.countplot(x='quality', data=Whitewine, label='White Wine')
plt.title('Comparison of Wine Quality', alpha=0.8)
plt.xlabel('Quality')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Create a box plot to compare the distribution of fixed acidity in red and white wines
plt.figure(figsize=(10,6))
sns.boxplot(x='wine_type', y='fixed acidity', data=pd.concat([Redwine.assign(wine_type='Red'), Whitewine.assign(wine_type='White')]))
plt.title('Comparison of Acidity Distribution')
plt.xlabel('Wine Type')
plt.ylabel('Acidity')
plt.show()
```

```python
# Create a violin plot to compare the distribution of pH in red and white wines
plt.figure(figsize=(12,8))
sns.catplot(x='wine_type', y='pH', data=pd.concat([Redwine.assign(wine_type='Red'), Whitewine.assign(wine_type='White')]))
plt.title('Comparison of pH Distribution')
plt.xlabel('Wine Type')
plt.ylabel('pH')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Assuming you have the Redwine dataframe

# Filter good and bad wines
good_wine = Redwine[Redwine['quality'] > 6]
bad_wine = Redwine[Redwine['quality'] <= 6]

# Create a single dataframe for histplot
wine_data = pd.concat([good_wine.drop('quality', axis=1), bad_wine.drop('quality', axis=1)], keys=['Good Wine', 'Bad Wine'])

# Create the histplot
sns.histplot(data=wine_data, multiple='stack', stat='density')

# Customize the plot (optional)
plt.xlabel('Component')
plt.ylabel('Density')  # Using density instead of mean value
plt.title('Distribution of Wine Components (Good vs. Bad)')
plt.legend(title='Wine Quality')
plt.show()
```

```python
good_wine = Redwine[Redwine['quality'] > 6]
bad_wine = Redwine[Redwine['quality'] <= 6]
mean_values_good = good_wine.drop('quality', axis=1).mean()
mean_values_bad = bad_wine.drop('quality', axis=1).mean()
plt.figure(figsize=(10, 6))
plt.bar(mean_values_good.index, mean_values_good.values, label='Good Wine')
plt.bar(mean_values_bad.index, mean_values_bad.values, label='Bad Wine')
plt.xlabel('Component')
plt.ylabel('Mean Value')
plt.title('Comparison of Mean Values for Good and Bad Wines')
plt.legend()
plt.show()
```
