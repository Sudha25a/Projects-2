```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
Redwine = pd.read_csv("winequality-red.csv", delimiter=';')
Redwine
```

```python
Whitewine = pd.read_csv("winequality-white.csv", delimiter=';')
Whitewine
```

```python
# Create a scatter plot to visualize the relationship between pH and fixed acidity
plt.figure(figsize=(10,6))
sns.scatterplot(x='pH', y='fixed acidity', data=Redwine, label='Red Wine')
sns.scatterplot(x='pH', y='fixed acidity', data=Whitewine, label='White Wine')
plt.title('Relationship between pH and Acidity')
plt.xlabel('pH')
plt.ylabel('fixed acidity')
plt.legend()
plt.show()
```

```python
# Create a bar plot to compare the quality of red and white wines
plt.figure(figsize=(10,6))
sns.countplot(x='quality', data=Redwine, label='Red Wine')
sns.countplot(x='quality', data=Whitewine, label='White Wine')
plt.title('Comparison of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# Create a box plot to compare the distribution of fixed acidity in red and white wines
plt.figure(figsize=(10,6))
sns.boxplot(x='wine_type', y='fixed acidity', data=pd.concat([Redwine.assign(wine_type='Red'), Whitewine.assign(wine_type='White')]))
plt.title('Comparison of Acidity Distribution')
plt.xlabel('Wine Type')
plt.ylabel('Acidity')
plt.show()
```

```python
# Create a violin plot to compare the distribution of pH in red and white wines
plt.figure(figsize=(10,6))
sns.catplot(x='wine_type', y='pH', data=pd.concat([Redwine.assign(wine_type='Red'), Whitewine.assign(wine_type='White')]))
plt.title('Comparison of pH Distribution')
plt.xlabel('Wine Type')
plt.ylabel('pH')
plt.show()
```

```python
# Select the continuous variable (e.g., sugar content)
variable = 'residual sugar'

# Create a histogram with 20 bins
sns.histplot(Redwine[variable], bins=20)
sns.histplot(Whitewine[variable], bins=20)
```

```python
#https://seaborn.pydata.org/search.html?q=kdeplot#

sns.kdeplot(Redwine[variable])
# sns.kdeplot(Redwine[variable], shade=True)
sns.kdeplot(Whitewine[variable])
# sns.kdeplot(Whitewine[variable], shade=True)

plt.title(f'Density Plot of {variable}')
plt.xlabel(variable)
plt.ylabel('Density')

# Show the plot
plt.show()
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
Redwine['quality_label'] = Redwine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into a categorical data type (specific to pandas) instead of a simple string
Redwine['quality_label'] = pd.Categorical(Redwine['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
Whitewine['quality_label'] = Whitewine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into a categorical data type (specific to pandas) instead of a simple string
Whitewine['quality_label'] = pd.Categorical(Whitewine['quality_label'],
categories=['low', 'medium', 'high'])
```
