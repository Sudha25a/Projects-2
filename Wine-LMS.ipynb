```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
Redwine['quality_label'] = Redwine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into a categorical data type (specific to pandas) instead of a simple string
Redwine['quality_label'] = pd.Categorical(Redwine['quality_label'],
categories=['low', 'medium', 'high'])
Redwine
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
Whitewine['quality_label'] = Whitewine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into a categorical data type (specific to pandas) instead of a simple string
Whitewine['quality_label'] = pd.Categorical(Whitewine['quality_label'],
categories=['low', 'medium', 'high'])
Whitewine
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 8), sharey=True)
sns.histplot(x='quality_label', data=Redwine, label='Red Wine', ax=axes[1], color='r')
sns.histplot(x='quality_label', data=Whitewine, label='White Wine', ax=axes[0], color='g')
axes[1].set_title('Red Wine Quality', alpha=0.2)
axes[0].set_title('White Wine Quality', alpha=0.2)
plt.show()
```

```python
# Concatenate the two DataFrames vertically
Winedata = pd.concat([Redwine, Whitewine])

Winedata
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(Redwine['residual sugar'], bins=10, color='red', edgecolor='black')
axes[0].set_title('Red Wine')
axes[0].set_xlabel('Residual Sugar')
axes[0].set_ylabel('Frequency')

axes[1].hist(Whitewine['residual sugar'], bins=10, color='white', edgecolor='black')
axes[1].set_title('White Wine')
axes[1].set_xlabel('Residual Sugar')
axes[1].set_ylabel('Frequency')

fig.suptitle('Residual Sugar Content in Wine')
plt.tight_layout()
plt.show()
```

Epic 2: Multivariate Analysis

```python
Whitewine = pd.read_csv("winequality-white.csv", delimiter=';')
Whitewine

# Create a correlation matrix for red wine features
plt.figure(figsize=(12,8))
corr_matrix= Whitewine.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True, mask=mask)
plt.title('White Wine Correlation Matrix')
plt.show()
```

```python
Redwine = pd.read_csv("winequality-red.csv", delimiter=';')
Redwine
# Create a correlation matrix for red wine features
plt.figure(figsize=(12,8))
corr_matrix_red = Redwine.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix_red, dtype=bool))
sns.heatmap(corr_matrix_red, annot=True, cmap='coolwarm', square=True, mask=mask)
plt.title('Red Wine Correlation Matrix')
plt.show()
```
