```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
Whitewine = pd.read_csv("winequality-white.csv", delimiter=';')
Whitewine
```

```python
plt.scatter(data=Whitewine, x= 'quality', y = 'alcohol', label='White Wine', color= '.2')
plt.title('Relationship between Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.legend()
plt.show()
```

```python
sns.barplot(data=Whitewine, x= 'quality', y = 'residual sugar', label='White Wine', color= '.2')
plt.title('Relationship between Quality and Residual sugar')
plt.xlabel('Quality')
plt.ylabel('residual sugar')
plt.legend()
plt.show()
```
